```python
# Ce programme calcule une prime d’assurance.
# L’âge du vehicule, la région d’origine du véhicule,
# l’âge du conducteur, ainsi que le total des
# réclamation au cours des 4 dernières années sont requis.

RISQUE_VEHICULE_ASIE = 1.5
RISQUE_VEHICULE_EUROPE = 1
RISQUE_VEHICULE_AMERIQUE = 0.75
AGE_INF = 30
AGE_SUP = 50
AGE_VIEUX_VEHICULE = 10

prime_assurance = 0
taux_risque = 0
risque_véhicule = 0
risque_conducteur = 0
risque_véhicule_origine = 0
risque_véhicule_âge = 0
risque_conducteur_réclamations = 0
risque_conducteur_âge = 0

age_du_vehicule = 0

# 1 si Amérique, 2 si Europe, 3 si Asie
region_origine_vehicule = 0
age_du_conducteur = 0

# total des réclamation au cours des 4 dernières années
total_réclamations = 0

# Logique
age_du_vehicule = 8
region_origine_vehicule = 2
age_du_conducteur = 35
total_réclamations = 200

if region_origine_vehicule == 1 :
    risque_véhicule_origine = RISQUE_VEHICULE_AMERIQUE
elif region_origine_vehicule == 2 :
    risque_véhicule_origine = RISQUE_VEHICULE_EUROPE
elif region_origine_vehicule == 3 :
    risque_véhicule_origine = RISQUE_VEHICULE_ASIE

if age_du_vehicule > AGE_VIEUX_VEHICULE :
    risque_véhicule_âge = 1
else :
    risque_véhicule_âge = 1 + ((AGE_VIEUX_VEHICULE+1) - age_du_vehicule) * 0.2

risque_véhicule = risque_véhicule_origine * risque_véhicule_âge

risque_conducteur_réclamations = total_réclamations / 1000 * 0.1

if age_du_conducteur > AGE_SUP :
    risque_conducteur_âge = 1 + (age_du_conducteur - AGE_SUP) * 0.1
elif age_du_conducteur < AGE_INF :
    risque_conducteur_âge = 1 + (AGE_INF - age_du_conducteur) * 0.2
else :
    risque_conducteur_âge = 1

risque_conducteur = risque_conducteur_réclamations + risque_conducteur_âge

taux_risque = risque_véhicule + risque_conducteur
prime_assurance = taux_risque * 150

if age_du_vehicule > 1 :
    print("L'âge du véhicule : {0} ans".format(age_du_vehicule))
else :
    print("L'âge du véhicule : {0} an".format(age_du_vehicule))

print("La région d'origine :")
if region_origine_vehicule == 1 :
    print("Amérique")
elif region_origine_vehicule == 2 :
    print("Europe")
elif region_origine_vehicule == 3 :
    print("Asie")

print("L'âge du conducteur : {0}".format(age_du_conducteur))
print("Le total des réclamation au cours des 4 dernières annéees : {0}".format(total_réclamations))
print("La prime d’assurance est {0}".format(prime_assurance))
```

```python
def produit_acheté(code_prod, quantité, taxable, prix_unitaire) :
    prod = {}
    prod["code_prod"] = code_prod
    prod["quantité"] = quantité
    prod["taxable"] = taxable
    prod["prix_unitaire"] = prix_unitaire
    return prod

def ajoute_produit(produit, liste) :
    # On cherche le produit dans la liste
    for i_produit in range(len(liste)) :
        # Si le produit est là, on ajoute la quantité, puis
        # on termine la méthode.
        if produit["code_prod"] == liste[i_produit]["code_prod"] :
            liste[i_produit]["quantité"] = liste[i_produit]["quantité"] + produit["quantité"]
            return
    # On ajoute le produit si on ne l’a pas trouvé dans la liste
    liste.append(produit)

def somme_taxables(liste) :
    total = 0
    for i_produit in liste :
        if i_produit["taxable"] :
            total = total + i_produit["quantité"] * i_produit["prix_unitaire"]
    return total
```

```python
def vérifier_expression(expression) :
    p = []
    bien_parenthésé = True
    for i_char in range(len(expression)) :
        caractère = expression[i_char]
        if caractère == "(" :
            p.append(caractère)
        elif caractère == ")" :
            if len(p) == 0 :
                bien_parenthésé = False
                break
            else :
                p.pop()
    return len(p) == 0
```

```python
# Implémentation d’un arbre binaire par chaînage. Un arbre est un noeud qui
# a possiblement des fils qui sont eux aussi des arbres.
# Constructeur. obj : objet auquel fait référence ce noeud de l’arbre.
def arbre_binaire (obj) :
    arbre = {}
    arbre["obj"] = obj
    arbre["gauche"] = None
    arbre["droite"] = None
    return arbre

# Calcule la taille de l’arbre. Méthode récursive qui calcule
# la taille des fils puis ajoute 1 à la somme des tailles
# des deux fils.
# Retourne la taille de l’arbre à partir du noeud courant.
def taille(arbre) :
    t = 1
    if arbre["gauche"] is not None :
        t = t + taille(arbre["gauche"])
    if arbre["droite"] is not None :
        t = t + taille(arbre["droite"])
    return t

# Calcule la hauteur de l’arbre. Méthode récursive qui calcule
# la hauteur des fils puis ajoute 1 à la plus grande hauteur
# parmis les deux fils.
# Retourne la hauteur de l’arbre à partir du noeud courant.
def hauteur(arbre) :
    h = 0
    if arbre["gauche"] is not None :
        h = 1+hauteur(arbre["gauche"])
    if arbre["droite"] is not None :
        h_droite = 1+hauteur(arbre["droite"])
        if h < h_droite :
            h = h_droite
    return h

# Recherche une valeur dans un arbre et retourne une
# référence au noeud contenant cette valeur.
# val est une référence à l’objet recherché.
# égal est une fonction comparant deux objets.
# Retourne la référence au premier noeud (visite en préordre) contenant
# la valeur recherchée.
def trouve (arbre, val, égal) :
    n = None
    if égal(arbre["obj"], val) :
        n = arbre
    if n is None and arbre["gauche"] is not None :
        n = trouve(arbre["gauche"], val, égal)
    if n is None and arbre["droite"] is not None :
        n = trouve(arbre["droite"], val, égal)
    return n

# Détermine si une valeur se trouve dans un arbre ou non.
# val est une référence à l’objet recherché.
# égal est une fonction comparant deux objets.
# Retourne True si l’objet se trouve dans l’arbre et False sinon.
def membre (arbre, val, égal) :
    trouvé = égal(arbre["obj"], val)
    if not trouvé and arbre["gauche"] is not None :
        trouvé = membre(arbre["gauche"], val, égal)
    if not trouvé and arbre["droite"] is not None :
        trouvé = membre(arbre["droite"], val, égal)
    return trouvé

# Visite en préordre d’un arbre. Cette méthode construit une chaîne de caractère qui
# est la concaténation des objets contenus dans les noeuds lorsque visités en préordre.
# Retourne les noeuds de l’arbre visité en préordre.
def préordre(arbre) :
    s = ""
    s = s+" "+str(arbre["obj"])
    if arbre["gauche"] is not None :
        s = s+préordre(arbre["gauche"])
    if arbre["droite"] is not None :
        s = s+préordre(arbre["droite"])
    return s

# Visite en postordre d’un arbre. Cette méthode construit une chaîne de caractère qui
# est la concaténation des objets contenus dans les noeuds lorsque visités en postordre.
# Retourne les noeuds de l’arbre visité en postordre.
def postordre(arbre) :
    s = ""
    if arbre["gauche"] is not None :
        s = s+postordre(arbre["gauche"])
    if arbre["droite"] is not None :
        s = s+postordre(arbre["droite"])
    s = s+" "+str(arbre["obj"])
    return s

# Visite en ordre d’un arbre. Cette méthode construit une chaîne de caractère qui
# est la concaténation des objets contenus dans les noeuds lorsque visités en ordre.
# Retourne les noeuds de l’arbre visité en ordre.
def enordre(arbre) :
    s = ""
    if arbre["gauche"] is not None :
        s = s+enordre(arbre["gauche"])
    s = s+" "+str(arbre["obj"])
    if arbre["droite"] is not None :
        s = s+enordre(arbre["droite"])
    return s
```

```python
# Méthode récursive d’insertion dans un arbre binaire de recherche.
# Elle garantit que la structure de l’arbre est préservée.
# noeud : racine du sous-arbre courant.
# val : la valeur à insérer.
# Retourne la nouvelle racine du sous-arbre après insertion.

def insère(noeud, val) :
    if noeud is not None :
        if val < noeud["obj"] :
            noeud["gauche"] = insère(noeud["gauche"],val)
        else :
            noeud["droite"] = insère(noeud["droite"],val)
    else :
        noeud = arbre_binaire(val)
    return noeud

# Méthode récursive de suppression dans un arbre binaire de recherche.
# Elle garantit que la structure de l’arbre est préservée.
# noeud : racine du sous-arbre courant.
# val : la valeur à supprimer.
# Retourne la nouvelle racine du sous-arbre après suppression.

def supprime(noeud, val) :
    n = noeud
    gauche = noeud["gauche"]
    droite = noeud["droite"]
    obj = noeud["obj"]

    if n is not None :
        if val < obj :
            noeud["gauche"] = supprime(gauche,val)
        elif val > obj :
            noeud["droite"] = supprime(droite,val)
        elif gauche is not None and droite is not None :
            n = gauche
            fils = gauche
            while fils["droite"] is not None :
                fils = fils["droite"]
            fils["droite"] = droite
        elif gauche is not None :
            n = gauche
        else :
            n = droite
    return n

# Algorithme qui détermine si un arbre (ou un sous-arbre) est un arbre binaire de recherche.

def est_arbre_binaire_recherche(arbre) :
    ok = True

    if arbre["gauche"] is not None :
        if trouve_max(arbre["gauche"]) > arbre["obj"] :
            ok = False
        else :
            ok = est_arbre_binaire_recherche(arbre["gauche"])
    if ok and arbre["droite"] is not None :
        if trouve_min(arbre["droite"]) < arbre["obj"] :
            ok = False
        else :
            ok = est_arbre_binaire_recherche(arbre["droite"])
    return ok

# Algorithme qui trouve la plus petite valeur dans un arbre.

def trouve_min(arbre) :
    min = arbre["obj"]
    if arbre["gauche"] is not None :
        min_sous_arbre = trouve_min (arbre["gauche"])
        if min > min_sous_arbre :
            min = min_sous_arbre
    if arbre["droite"] is not None :
        min_sous_arbre = trouve_min (arbre["droite"])
        if min > min_sous_arbre :
            min = min_sous_arbre
    return min

# Algorithme qui trouve la plus petite valeur dans un arbre.
def trouve_max(arbre) :
    max = arbre["obj"]
    if arbre["gauche"] is not None :
        max_sous_arbre = trouve_max (arbre["gauche"])
        if max < max_sous_arbre :
            max = max_sous_arbre
    if arbre["droite"] is not None :
        max_sous_arbre = trouve_max (arbre["droite"])
        if max < max_sous_arbre :
            max = max_sous_arbre
    return max
```

```python
from collections import deque

# Le caractère A utilisé pour afficher le nom des noeuds du graphe.

CHAR_A = 65

# Constructeur.
# n : le nombre de sommets dans le graphe.
def graphe (n) :
    return [ [ 0 for j in range(n) ] for i in range(n) ]

# Permet d’ajouter une arête entre deux sommets i et j.
# i : le premier sommet.
# j : le deuxième sommet.

def ajouter_arête(g, i, j) :
    if g[i][j] == 0 :
        g[i][j]=1
        g[j][i]=1

# Permet de supprimer une arête entre deux sommets i et j.
# i : le premier sommet.
# j : le deuxième sommet.

def retirer_arête(g, i, j) :
    if g[i][j] != 0 :
        g[i][j]=0
        g[j][i]=0

# Détermine si les sommets i et j sont adjacents.
# i : le premier sommet.
# j : le deuxième sommet.
# Retourne True si les sommets sont adjacents ; False, autrement.

def adjacent(g, i, j) :
    return g[i][j] !=0

# Détermine le degré d’un sommet.
# i : le sommet pour lequel on calcule le degré.
# Retourne le nombre de sommets adjacents au sommet i.

def degré(g, i) :
    d=0
    for j in range(len(g)) :
        if g[i][j] != 0 :
            d = d + 1
    return d

# Pour le parcours en profondeur. Cette méthode est une auxiliaire
# pour la méthode parcourir_en_profondeur. Elle retourne le parcours
# en profondeur à partir de début vers les sommets non déjà visités qui
# sont indiqués dans visité.
# début : le sommet présentement visité.
# visité : liste des sommets visités jusqu’à présent.
# Retourne la liste des noeuds visités dans l’ordre de visite.

def avancer(g, début, visité) :
    visité.append(début)

    # puis on avance d’un niveau
    for k in range(len(g)) :
        if k not in visité and adjacent(g, début, k) :
            avancer(g, k, visité)

# Visite en profondeur du graphe à partir d’un sommet.
# s : le sommet de départ.
# Retourne la liste des noeuds visités dans l’ordre de visite.

def parcourir_en_profondeur (g, départ) :
    visité = []

    avancer(g, départ, visité)

    return visité

# Visite en largeur du graphe à partir d’un sommet.
# départ : le sommet de départ.
# Retourne la liste des noeuds visités dans l’ordre de visite.

def parcourir_en_largeur(g, départ) :
    f = deque()
    visité = []

    f.append(départ)
    while len(f) != 0 :
        départ=f.popleft()
        if départ not in visité :
            visité.append(départ)
            for k in range(len(g)) :
                if k not in visité and adjacent(g, départ, k) :
                    f.append(k)
    return visité
```

```python
g = graphe(9)

ajouter_arête(g, 0, 3) # A - D
ajouter_arête(g, 0, 5) # A - F
ajouter_arête(g, 1, 2) # B - C
ajouter_arête(g, 1, 3) # B - D
ajouter_arête(g, 1, 7) # B - H
ajouter_arête(g, 2, 3) # C - D
ajouter_arête(g, 2, 8) # C - I
ajouter_arête(g, 2, 7) # C - H
ajouter_arête(g, 3, 5) # D - F
ajouter_arête(g, 4, 5) # E - F
ajouter_arête(g, 5, 6) # F - G
ajouter_arête(g, 5, 7) # F - H
ajouter_arête(g, 6, 7) # G - H
ajouter_arête(g, 6, 8) # G - I

print(g)
```

```python
out = ""
for s in parcourir_en_profondeur(g, 0):
    out = out + chr(CHAR_A + s) + " "
print(out)

out = ""
for s in parcourir_en_largeur(g, 0):
    out = out + chr(CHAR_A + s) + " "
print(out)
```

```python
import sys

# Détermine le plus court chemin à partir de l’origine en
# appliquant l’algorithme de Dijkstra.
# g : le graphe.
# départ : sommet d’origine.
# coûts : tableau des coûts optimaux.
# transits : tableau des noeuds précédents.

def plus_court_chemin(g, départ, coûts, transits) :
    # Algorithme pour trouver tous les chemins à moindre coût d’un noeud
    # à tous les autres noeuds (algorithme de Dijktra)

    min_val = sys.maxsize
    courant = départ

    sommets_restants = []
    for s in range(len(g)) :
        if s != départ :
            sommets_restants.insert(0,s)
            if adjacent(g, départ,s) :
                coûts[s] =  g[départ][s]
                transits[s] = départ
            else :
                coûts[s] = sys.maxsize
                transits[s] = -1
        else :
            coûts[s] = 0
            transits[s] = départ
    while len(sommets_restants) != 0 :
        for s in sommets_restants :
            if adjacent(g, courant,s) and coûts[s] > coûts[courant]+g[courant][s] :
                coûts[s] = coûts[courant]+g[courant][s]
                transits[s] = courant
        courant = sommets_restants[0]
        min_val = coûts[courant]
        for s in sommets_restants :
            if coûts[s] < min_val :
                courant = s
                min_val = coûts[courant]
        sommets_restants.remove(courant)
```

```python
coûts = [0 for i in range(len(g))]
transits = [-1 for i in range(len(g))]

plus_court_chemin(g, 2, coûts, transits)

print("Table à partir du noeud {0}".format(chr(CHAR_A+2)))
print("Noeud - Précédent - Coût")
for i in range(len(g)) :
    print("  {0}   -     {1}     - {2}".format(chr(CHAR_A+i), chr(CHAR_A+transits[i]), coûts[i]))
```
